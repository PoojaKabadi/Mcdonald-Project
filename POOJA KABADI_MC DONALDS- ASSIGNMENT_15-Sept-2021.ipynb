# Q4 - Which category contributes to the maximum % of Cholesterol in a diet (% daily value)?

```python
a = pd.pivot_table(df,'Cholesterol (% Daily Value)', index= ['Category'])
a.sort_values(('Cholesterol (% Daily Value)'), ascending= False).reset_index()
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x='Category', y='Cholesterol (% Daily Value)', data=df, ci=None, estimator=np.sum);
plt.xticks(rotation=90)
plt.show()
```

### Inference
 - Breakfast category contributes to the maximum % of of Cholesterol in a diet(% daily value)
 - The above table is sorted and arranged the Categories from the High to Low values based on the Cholesterol(% Daily Value)

# Q5 - Which item contributes maximum to the Sodium intake?
